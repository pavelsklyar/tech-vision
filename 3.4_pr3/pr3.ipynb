```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, util, filters, metrics
import cv2
from prettytable import PrettyTable 
```

```python
freq= 0.00366772681 * 100
phase=0

def function(x, chastota, ph): #  
    return(0.5*np.sin(x*chastota+ph)+0.5)

def function_dis(freq, phase): #  
    out=np.array([])
    for i in np.arange(0,freq*3000,freq):
        out=np.append(out,function(i, freq, phase))
        # print(i)
    return out

def draw(frequency, phase):#   
    x=np.arange(0,frequency*3000,frequency)
    plt.plot(x,function(x, frequency, phase))
    # plt.xlim(0,300)
    plt.ylim(0,1)
    plt.show()

draw(freq, phase)
draw(0.2 * freq, phase)
draw(freq / 3, phase)
draw(0.5 * freq, phase)
draw(1 * freq, phase)
draw(1.41 * freq, phase)
draw(2 * freq, phase)
draw(3 * freq, phase)
```

```python
#     
def image(row): 
    row = np.around(255 * row)
    image = np.full((row.size, row.size), row)
    return image
```

```python

img1 = image(function_dis(0.2 * freq, phase))
img2 = image(function_dis(freq / 3, phase))
img3 = image(function_dis(0.5 * freq, phase))
img4 = image(function_dis(1 * freq, phase))
img5 = image(function_dis(1.41 * freq, phase))
img6 = image(function_dis(2 * freq, phase))
img7 = image(function_dis(3 * freq, phase))

fig = plt.figure()

fig, ([ax1, ax2, ax3, ax4, ax5, ax6,ax7] ) = plt.subplots(
    nrows=1, ncols=7,
    figsize=(100, 100)
)

ax1.set_title('01')
ax2.set_title('02')
ax3.set_title('03')
ax4.set_title('04')
ax5.set_title('05')
ax6.set_title('06')
ax7.set_title('07')


ax1.imshow(img1, cmap='gray')
ax2.imshow(img2, cmap='gray')
ax3.imshow(img3, cmap='gray')
ax4.imshow(img4,cmap='gray')
ax5.imshow(img5, cmap='gray')
ax6.imshow(img6, cmap='gray')
ax7.imshow(img7,cmap='gray')

plt.show()
```
